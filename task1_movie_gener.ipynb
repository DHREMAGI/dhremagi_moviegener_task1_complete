```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,r2_score
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
```

```python
encode=LabelEncoder()
combined_dataset['Gener'] = encode.fit_transform(combined_dataset['Gener'].values)
```

```python
encode=LabelEncoder()
combined_dataset['movie_name'] = encode.fit_transform(combined_dataset['movie_name'].values)
```

```python
combined_dataset.head(10)
```

```python
combined_dataset.Gener = combined_dataset.Gener.fillna(combined_dataset.Gener.mean())
combined_dataset.count()
```

```python
combined_dataset.duplicated().values.any()
```

```python
combined_dataset=combined_dataset.drop_duplicates()
```

```python
combined_dataset.duplicated().values.any()
```

PROCESS THE DATA

```python
vectorizer = TfidfVectorizer()
x = vectorizer.fit_transform(combined_dataset['Description'])
combined_dataset.head(3)
```

```python
y = combined_dataset['Gener']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

TRAIN NAIVE BAYES CLASSIFIER

```python
from sklearn.naive_bayes import MultinomialNB
naive_bayes_model = MultinomialNB()
naive_bayes_model.fit(x_train, y_train)
```

```python
data_predictions = naive_bayes_model.predict(x_test)
```

```python
print("NAIVE BAYES CLASSIFIER MODEL")
print(confusion_matrix(y_test, data_predictions))
print(classification_report(y_test, data_predictions))
print("Accuracy: ",accuracy_score(y_test, data_predictions))
print("r2_score: ",r2_score(y_test, data_predictions))
```

TRAIN LOGISTIC REGRESSION

```python
logistic_regression_model = LogisticRegression()
```

```python
logistic_regression_model.fit(x_train,y_train)
```

```python
logr_predictions = logistic_regression_model.predict(x_test)

print("LOGISTIC REGRESSION MODEL")
print(confusion_matrix(y_test, logr_predictions))
print(classification_report(y_test, logr_predictions))
print("Accuracy:",accuracy_score(y_test, logr_predictions))
print("r2_score:",r2_score(y_test,logr_predictions))
```
